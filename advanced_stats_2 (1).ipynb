```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Question 1: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
def z_test_one_sample(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test and interprets the results.
    Args:
        sample_data (array-like): The sample observations.
        pop_mean (float): The known population mean under the null hypothesis.
        pop_std (float): The known population standard deviation.
        alpha (float): The significance level (default is 0.05).
    Returns:
        tuple: (z_score, p_value, interpretation_message)
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Calculate p-value (two-tailed)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpret results
    if p_value < alpha:
        message = f"Reject the null hypothesis. There is a statistically significant difference between the sample mean ({sample_mean:.2f}) and the population mean ({pop_mean})."
    else:
        message = f"Fail to reject the null hypothesis. There is no statistically significant difference between the sample mean ({sample_mean:.2f}) and the population mean ({pop_mean})."

    return z_score, p_value, message

# Example Usage for Q1:
np.random.seed(42)
sample_data_q1 = np.random.normal(loc=52, scale=10, size=50) # Sample from a population with mean 52
pop_mean_q1 = 50 # Hypothesized population mean
pop_std_q1 = 10 # Known population standard deviation

z_score_q1, p_value_q1, interpretation_q1 = z_test_one_sample(sample_data_q1, pop_mean_q1, pop_std_q1)
print(f"Q1 Z-Test Results:")
print(f"Z-Score: {z_score_q1:.3f}")
print(f"P-Value: {p_value_q1:.3f}")
print(f"Interpretation: {interpretation_q1}\n")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Question 4: Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
def two_tailed_z_test_visual(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision regions.
    Args:
        sample_data (array-like): The sample observations.
        pop_mean (float): The known population mean under the null hypothesis.
        pop_std (float): The known population standard deviation.
        alpha (float): The significance level (default is 0.05).
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject H0 (p={p_value:.3f} < alpha={alpha}). Sample mean is significantly different from population mean."
    else:
        interpretation = f"Fail to reject H0 (p={p_value:.3f} >= alpha={alpha}). No significant difference."

    # Calculate critical Z-values for two tails
    critical_z_lower = stats.norm.ppf(alpha / 2)
    critical_z_upper = stats.norm.ppf(1 - alpha / 2)

    # Plotting the Z-distribution
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')

    # Shade rejection regions
    x_rejection_lower = x[x < critical_z_lower]
    x_rejection_upper = x[x > critical_z_upper]
    plt.fill_between(x_rejection_lower, 0, stats.norm.pdf(x_rejection_lower), color='red', alpha=0.5, label='Rejection Region (alpha/2)')
    plt.fill_between(x_rejection_upper, 0, stats.norm.pdf(x_rejection_upper), color='red', alpha=0.5)

    # Plot calculated Z-score
    plt.axvline(z_score, color='green', linestyle='--', label=f'Calculated Z-score: {z_score:.2f}')

    plt.title(f'Two-Tailed Z-Test Visualization (alpha = {alpha})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.axvline(critical_z_lower, color='black', linestyle=':', label=f'Critical Z: {critical_z_lower:.2f}, {critical_z_upper:.2f}')
    plt.axvline(critical_z_upper, color='black', linestyle=':')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Q4 Two-Tailed Z-Test Visualization Results:")
    print(f"Z-Score: {z_score:.3f}, P-Value: {p_value:.3f}")
    print(f"Interpretation: {interpretation}\n")

# Example Usage for Q4:
np.random.seed(44)
sample_data_q4 = np.random.normal(loc=102, scale=12, size=60)
pop_mean_q4 = 100
pop_std_q4 = 12
two_tailed_z_test_visual(sample_data_q4, pop_mean_q4, pop_std_q4)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Question 5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type_errors(pop_mean_null, pop_std, sample_size, alpha, true_pop_mean_alt=None):
    """
    Visualizes Type I and Type II errors for a one-sample Z-test.
    Args:
        pop_mean_null (float): Population mean under the null hypothesis.
        pop_std (float): Known population standard deviation.
        sample_size (int): Sample size.
        alpha (float): Significance level for Type I error.
        true_pop_mean_alt (float, optional): True population mean under the alternative hypothesis.
                                            If None, only Type I error is shown.
    """
    std_error = pop_std / np.sqrt(sample_size)

    # Critical Z-value for alpha (two-tailed)
    critical_z = stats.norm.ppf(1 - alpha / 2)

    # Critical sample mean values
    critical_mean_lower = pop_mean_null - critical_z * std_error
    critical_mean_upper = pop_mean_null + critical_z * std_error

    plt.figure(figsize=(12, 7))

    # Plot Null Distribution
    x = np.linspace(pop_mean_null - 4 * std_error, pop_mean_null + 4 * std_error, 1000)
    pdf_null = stats.norm.pdf(x, loc=pop_mean_null, scale=std_error)
    plt.plot(x, pdf_null, label='Distribution under Null Hypothesis', color='blue')

    # Shade Type I Error (Alpha) region
    x_type1_lower = x[x < critical_mean_lower]
    x_type1_upper = x[x > critical_mean_upper]
    plt.fill_between(x_type1_lower, 0, stats.norm.pdf(x_type1_lower, loc=pop_mean_null, scale=std_error), color='red', alpha=0.3, label='Type I Error (Alpha)')
    plt.fill_between(x_type1_upper, 0, stats.norm.pdf(x_type1_upper, loc=pop_mean_null, scale=std_error), color='red', alpha=0.3)

    plt.axvline(pop_mean_null, color='blue', linestyle='--', label='Null Hypo. Mean')
    plt.axvline(critical_mean_lower, color='red', linestyle=':', label=f'Critical Means (alpha={alpha})')
    plt.axvline(critical_mean_upper, color='red', linestyle=':')

    if true_pop_mean_alt is not None:
        # Plot Alternative Distribution
        x_alt = np.linspace(true_pop_mean_alt - 4 * std_error, true_pop_mean_alt + 4 * std_error, 1000)
        pdf_alt = stats.norm.pdf(x_alt, loc=true_pop_mean_alt, scale=std_error)
        plt.plot(x_alt, pdf_alt, label='Distribution under Alternative Hypothesis', color='green')
        plt.axvline(true_pop_mean_alt, color='green', linestyle='--', label='True Alt. Mean')

        # Type II Error region (Beta)
        # This is the area under the alternative distribution that falls within the acceptance region of the null.
        type_ii_error_prob = stats.norm.cdf(critical_mean_upper, loc=true_pop_mean_alt, scale=std_error) - \
                             stats.norm.cdf(critical_mean_lower, loc=true_pop_mean_alt, scale=std_error)

        x_beta = np.linspace(critical_mean_lower, critical_mean_upper, 100)
        y_beta = stats.norm.pdf(x_beta, loc=true_pop_mean_alt, scale=std_error)
        plt.fill_between(x_beta, 0, y_beta, color='orange', alpha=0.5, label=f'Type II Error (Beta={type_ii_error_prob:.3f})')

        print(f"Q5 Type II Error (Beta): {type_ii_error_prob:.3f}")
        print(f"Power (1 - Beta): {1 - type_ii_error_prob:.3f}")

    plt.title('Type I (Alpha) and Type II (Beta) Error Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage for Q5:
print("\nQ5 Type I and Type II Error Visualization:")
visualize_type_errors(pop_mean_null=50, pop_std=10, sample_size=30, alpha=0.05, true_pop_mean_alt=53)
```

```python
import numpy as np
import scipy.stats as stats

# Question 6: Write a Python program to perform an independent T-test and interpret the results.
def independent_t_test(sample1, sample2, alpha=0.05, equal_variances=True):
    """
    Performs an independent samples T-test and interprets the results.
    Args:
        sample1 (array-like): Data for the first sample.
        sample2 (array-like): Data for the second sample.
        alpha (float): The significance level (default is 0.05).
        equal_variances (bool): If True, assumes equal population variances (Student's t-test).
                                If False, performs Welch's t-test (default in scipy).
    Returns:
        tuple: (t_statistic, p_value, interpretation_message)
    """
    # Use equal_var parameter to specify whether to assume equal variances
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_variances)

    if p_value < alpha:
        message = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the means of the two independent groups."
    else:
        message = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the means of the two independent groups."

    return t_statistic, p_value, message

# Example Usage for Q6:
np.random.seed(45)
group_a = np.random.normal(loc=70, scale=10, size=40)
group_b = np.random.normal(loc=75, scale=10, size=45)

t_stat_q6, p_value_q6, interpretation_q6 = independent_t_test(group_a, group_b)
print(f"\nQ6 Independent T-Test Results:")
print(f"T-Statistic: {t_stat_q6:.3f}")
print(f"P-Value: {p_value_q6:.3f}")
print(f"Interpretation: {interpretation_q6}\n")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Question 7: Perform a paired sample T-test using Python and visualize the comparison results.
def paired_t_test_visual(sample_before, sample_after, alpha=0.05):
    """
    Performs a paired samples T-test and visualizes the comparison.
    Args:
        sample_before (array-like): Data from the 'before' condition.
        sample_after (array-like): Data from the 'after' condition.
        alpha (float): The significance level (default is 0.05).
    Returns:
        tuple: (t_statistic, p_value, interpretation_message)
    """
    if len(sample_before) != len(sample_after):
        raise ValueError("Paired samples must have the same number of observations.")

    t_statistic, p_value = stats.ttest_rel(sample_before, sample_after)

    if p_value < alpha:
        message = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between 'before' and 'after' measurements."
    else:
        message = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between 'before' and 'after' measurements."

    # Visualization (Boxplots)
    plt.figure(figsize=(8, 6))
    data_to_plot = [sample_before, sample_after]
    labels = ['Before', 'After']
    plt.boxplot(data_to_plot, labels=labels, patch_artist=True, medianprops={'color': 'black'})
    plt.title('Paired Sample Comparison (Boxplots)')
    plt.ylabel('Value')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Visualization (Distribution of Differences)
    differences = np.array(sample_after) - np.array(sample_before)
    plt.figure(figsize=(8, 6))
    plt.hist(differences, bins=15, edgecolor='black', alpha=0.7)
    plt.axvline(np.mean(differences), color='red', linestyle='--', label=f'Mean Difference: {np.mean(differences):.2f}')
    plt.title('Distribution of Differences (After - Before)')
    plt.xlabel('Difference')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return t_statistic, p_value, message

# Example Usage for Q7:
np.random.seed(46)
# Simulate 'before' and 'after' measurements for the same subjects
blood_pressure_before = np.random.normal(loc=140, scale=10, size=30)
# Assume medication causes a slight reduction
blood_pressure_after = blood_pressure_before - np.random.normal(loc=5, scale=3, size=30)

t_stat_q7, p_value_q7, interpretation_q7 = paired_t_test_visual(blood_pressure_before, blood_pressure_after)
print(f"\nQ7 Paired Sample T-Test Results:")
print(f"T-Statistic: {t_stat_q7:.3f}")
print(f"P-Value: {p_value_q7:.3f}")
print(f"Interpretation: {interpretation_q7}\n")
```

```python
import numpy as np
import scipy.stats as stats

# Question 8: Simulate data and perform both Z-test and T-test, then compare the results using Python.
def compare_z_t_tests(pop_mean, pop_std, sample_size, hypothesized_mean):
    """
    Simulates data and performs both Z-test and T-test for comparison.
    Args:
        pop_mean (float): True population mean.
        pop_std (float): True population standard deviation.
        sample_size (int): Size of the simulated sample.
        hypothesized_mean (float): Mean to test against (null hypothesis).
    """
    simulated_data = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)

    # Z-test (using known population std)
    sample_mean_z = np.mean(simulated_data)
    z_score = (sample_mean_z - hypothesized_mean) / (pop_std / np.sqrt(sample_size))
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # T-test (using sample std)
    t_statistic, p_value_t = stats.ttest_1samp(simulated_data, hypothesized_mean)

    print(f"\nQ8 Comparison of Z-Test and T-Test (Sample Size: {sample_size}):")
    print(f"Simulated Data Mean: {np.mean(simulated_data):.3f}")
    print(f"Hypothesized Mean: {hypothesized_mean}")
    print(f"--- Z-Test ---")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value (Z-Test): {p_value_z:.3f}")
    print(f"--- T-Test ---")
    print(f"T-Statistic: {t_statistic:.3f}")
    print(f"P-Value (T-Test): {p_value_t:.3f}")

    if sample_size < 30:
        print("Observation: For smaller sample sizes, T-test p-value tends to be higher (more conservative) than Z-test.")
    else:
        print("Observation: For larger sample sizes, Z-test and T-test results are very similar as t-distribution approximates normal.")
    print("-" * 30)

# Example Usage for Q8:
np.random.seed(47)
compare_z_t_tests(pop_mean=60, pop_std=15, sample_size=20, hypothesized_mean=65) # Small sample
compare_z_t_tests(pop_mean=60, pop_std=15, sample_size=100, hypothesized_mean=65) # Large sample
```

```python
import numpy as np
import scipy.stats as stats

# Question 9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean.
    Args:
        sample_data (array-like): The sample observations.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
    Returns:
        tuple: (lower_bound, upper_bound)
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1) # Use ddof=1 for sample standard deviation
    n = len(sample_data)

    # Degrees of freedom for t-distribution (since population std is unknown)
    df = n - 1

    # Calculate critical t-value (for two-tailed interval)
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df)

    # Calculate Margin of Error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"\nQ9 Confidence Interval Significance:")
    print(f"A {confidence_level*100:.0f}% confidence interval means that if we were to take many samples and construct an interval for each, approximately {confidence_level*100:.0f}% of those intervals would contain the true population mean.")
    print(f"It provides a range of plausible values for the population mean, reflecting the uncertainty due to sampling variability.")

    return lower_bound, upper_bound

# Example Usage for Q9:
np.random.seed(48)
sample_data_q9 = np.random.normal(loc=80, scale=15, size=50)
lower_ci_q9, upper_ci_q9 = calculate_confidence_interval(sample_data_q9, confidence_level=0.95)
print(f"95% Confidence Interval for Sample Mean: ({lower_ci_q9:.2f}, {upper_ci_q9:.2f})\n")
```

```python
import numpy as np
import scipy.stats as stats

# Question 10: Write a Python program to calculate the margin of error for a given confidence level using sample data.
def calculate_margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculates the margin of error for a sample mean.
    Args:
        sample_data (array-like): The sample observations.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
    Returns:
        float: The calculated margin of error.
    """
    sample_std = np.std(sample_data, ddof=1) # Use ddof=1 for sample standard deviation
    n = len(sample_data)
    df = n - 1

    alpha = 1 - confidence_level

    # Critical value (t-score for unknown population std, large sample uses z-score)
    # Using t-distribution which is more general for unknown population std
    critical_value = stats.t.ppf(1 - alpha / 2, df)

    # Standard Error of the Mean
    standard_error = sample_std / np.sqrt(n)

    margin_of_error = critical_value * standard_error

    return margin_of_error

# Example Usage for Q10:
np.random.seed(49)
sample_data_q10 = np.random.normal(loc=25, scale=5, size=30)
margin_error_q10 = calculate_margin_of_error(sample_data_q10, confidence_level=0.95)
print(f"Q10 Margin of Error for 95% Confidence Level: {margin_error_q10:.3f}\n")
```

```python
import numpy as np

# Question 11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
# This is a simplified conceptual example for a discrete case.
# More complex Bayesian inference typically uses libraries like PyMC3 or Pyro for MCMC.

def bayes_theorem_example(prior_A, likelihood_B_given_A, prob_B):
    """
    Calculates the posterior probability P(A|B) using Bayes' Theorem.
    Args:
        prior_A (float): P(A) - Prior probability of hypothesis A.
        likelihood_B_given_A (float): P(B|A) - Likelihood of observing evidence B given A is true.
        prob_B (float): P(B) - Marginal probability of observing evidence B.
    Returns:
        float: P(A|B) - Posterior probability of A given B.
    """
    if prob_B == 0:
        raise ValueError("Probability of evidence P(B) cannot be zero.")

    # P(A|B) = [P(B|A) * P(A)] / P(B)
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

# Explanation of the process:
# 1. Define a Prior Probability (P(A)): This is our initial belief in the hypothesis (A) before observing any new evidence.
# 2. Define the Likelihood (P(B|A)): This is the probability of observing the new evidence (B) if our hypothesis (A) is true.
# 3. Calculate the Marginal Probability of Evidence (P(B)): This is the overall probability of observing the evidence B, considering all possible hypotheses. If A is the only hypothesis, P(B) = P(B|A)P(A) + P(B|not A)P(not A).
# 4. Apply Bayes' Theorem: Use the formula P(A|B) = [P(B|A) * P(A)] / P(B) to compute the Posterior Probability (P(A|B)).
#    The posterior probability is our updated belief in the hypothesis A after taking the new evidence B into account.

# Example Usage for Q11: Medical Diagnosis
# Scenario: A rare disease (A) affects 1% of the population. A test (B) is 90% accurate (P(B|A)=0.9)
# and has a 5% false positive rate (P(B|not A)=0.05). If a person tests positive, what's the probability they have the disease?

# 1. Prior Probability P(Disease):
prior_disease = 0.01 # 1% prevalence

# 2. Likelihood P(Positive Test | Disease):
likelihood_positive_given_disease = 0.90 # Test sensitivity

# 3. Likelihood P(Positive Test | No Disease) (False Positive Rate):
likelihood_positive_given_no_disease = 0.05

# 4. Prior Probability P(No Disease):
prior_no_disease = 1 - prior_disease # 99%

# 5. Marginal Probability P(Positive Test) (P(B)):
# P(Positive Test) = P(Positive Test | Disease)P(Disease) + P(Positive Test | No Disease)P(No Disease)
prob_positive_test = (likelihood_positive_given_disease * prior_disease) + \
                     (likelihood_positive_given_no_disease * prior_no_disease)

# Calculate Posterior Probability P(Disease | Positive Test)
posterior_disease_given_positive = bayes_theorem_example(
    prior_A=prior_disease,
    likelihood_B_given_A=likelihood_positive_given_disease,
    prob_B=prob_positive_test
)

print(f"Q11 Bayesian Inference Example:")
print(f"Prior probability of disease: {prior_disease:.4f}")
print(f"Likelihood of positive test given disease: {likelihood_positive_given_disease:.4f}")
print(f"Likelihood of positive test given no disease (false positive): {likelihood_positive_given_no_disease:.4f}")
print(f"Marginal probability of positive test: {prob_positive_test:.4f}")
print(f"Posterior probability of disease given a positive test: {posterior_disease_given_positive:.4f}\n")
print("Explanation: Even with a positive test, the probability of having the disease is still relatively low because the disease is very rare. Bayes' Theorem helps us update our initial belief with new evidence.")
```

```python
import pandas as pd
import scipy.stats as stats

# Question 12: Perform a Chi-square test for independence between two categorical variables in Python.
def chi_square_test_independence(data, var1_col, var2_col, alpha=0.05):
    """
    Performs a Chi-square test for independence between two categorical variables.
    Args:
        data (pd.DataFrame): DataFrame containing the categorical variables.
        var1_col (str): Name of the first categorical variable column.
        var2_col (str): Name of the second categorical variable column.
        alpha (float): Significance level.
    Returns:
        tuple: (chi2_statistic, p_value, df, expected_frequencies, interpretation_message)
    """
    # Create contingency table
    contingency_table = pd.crosstab(data[var1_col], data[var2_col])

    # Perform Chi-square test
    chi2_statistic, p_value, df, expected_frequencies = stats.chi2_contingency(contingency_table)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant association (dependence) between {var1_col} and {var2_col}."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant association (independence) between {var1_col} and {var2_col}."

    return chi2_statistic, p_value, df, expected_frequencies, interpretation

# Example Usage for Q12:
np.random.seed(50)
data_q12 = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=200),
    'Preference': np.random.choice(['Option A', 'Option B', 'Option C'], size=200)
})
# Introduce some dependence for demonstration
data_q12.loc[data_q12['Gender'] == 'Male', 'Preference'] = np.random.choice(['Option A', 'Option B'], size=data_q12['Gender'].sum())

chi2_stat_q12, p_value_q12, df_q12, expected_freq_q12, interpretation_q12 = \
    chi_square_test_independence(data_q12, 'Gender', 'Preference')

print(f"Q12 Chi-square Test for Independence:")
print(f"Contingency Table:\n{pd.crosstab(data_q12['Gender'], data_q12['Preference'])}")
print(f"Chi-square Statistic: {chi2_stat_q12:.3f}")
print(f"P-Value: {p_value_q12:.3f}")
print(f"Degrees of Freedom: {df_q12}")
print(f"Expected Frequencies:\n{expected_freq_q12}")
print(f"Interpretation: {interpretation_q12}\n")
```

```python
import numpy as np
import pandas as pd

# Question 13: Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
def calculate_expected_frequencies(observed_data, test_type='goodness_of_fit', hypothesized_proportions=None):
    """
    Calculates expected frequencies for Chi-square tests.
    Args:
        observed_data (array-like or pd.DataFrame): Observed frequencies (for goodness-of-fit)
                                                    or a contingency table (for independence).
        test_type (str): 'goodness_of_fit' or 'independence'.
        hypothesized_proportions (array-like, optional): For goodness-of-fit, the expected proportions.
                                                         Must sum to 1. Not used for 'independence'.
    Returns:
        np.array or pd.DataFrame: Calculated expected frequencies.
    """
    if test_type == 'goodness_of_fit':
        observed_frequencies = np.asarray(observed_data)
        total_observations = np.sum(observed_frequencies)

        if hypothesized_proportions is None:
            # Assume equal distribution if proportions not given
            hypothesized_proportions = np.ones_like(observed_frequencies) / len(observed_frequencies)

        if not np.isclose(np.sum(hypothesized_proportions), 1.0):
            raise ValueError("Hypothesized proportions must sum to 1.")

        expected_frequencies = total_observations * hypothesized_proportions
        return expected_frequencies

    elif test_type == 'independence':
        if not isinstance(observed_data, pd.DataFrame):
            raise TypeError("For 'independence' test, observed_data should be a pandas DataFrame (contingency table).")

        row_totals = observed_data.sum(axis=1)
        col_totals = observed_data.sum(axis=0)
        grand_total = observed_data.sum().sum()

        expected_frequencies = pd.DataFrame(index=observed_data.index, columns=observed_data.columns)
        for i in observed_data.index:
            for j in observed_data.columns:
                expected_frequencies.loc[i, j] = (row_totals[i] * col_totals[j]) / grand_total
        return expected_frequencies

    else:
        raise ValueError("test_type must be 'goodness_of_fit' or 'independence'.")

# Example Usage for Q13:
# Goodness-of-Fit
observed_q13_gof = np.array([25, 35, 40]) # e.g., counts of colors in a bag of candies
hypo_prop_q13_gof = np.array([0.2, 0.3, 0.5]) # e.g., claimed proportions by manufacturer
expected_freq_q13_gof = calculate_expected_frequencies(observed_q13_gof, 'goodness_of_fit', hypo_prop_q13_gof)
print(f"Q13 Expected Frequencies (Goodness-of-Fit):\n{expected_freq_q13_gof}\n")

# Independence
data_q13_ind = pd.DataFrame({
    'Gender': ['M','M','M','F','F','F'],
    'Opinion': ['Yes','No','Yes','No','Yes','No']
})
contingency_table_q13_ind = pd.crosstab(data_q13_ind['Gender'], data_q13_ind['Opinion'])
expected_freq_q13_ind = calculate_expected_frequencies(contingency_table_q13_ind, 'independence')
print(f"Q13 Expected Frequencies (Independence):\n{expected_freq_q13_ind}\n")
```

```python
import numpy as np
import scipy.stats as stats

# Question 14: Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
def chi_square_goodness_of_fit(observed_frequencies, expected_frequencies=None, alpha=0.05):
    """
    Performs a Chi-square goodness-of-fit test.
    Args:
        observed_frequencies (array-like): The observed counts for each category.
        expected_frequencies (array-like, optional): The expected counts for each category.
                                                      If None, assumes equal expected frequencies.
        alpha (float): Significance level.
    Returns:
        tuple: (chi2_statistic, p_value, interpretation_message)
    """
    observed_frequencies = np.asarray(observed_frequencies)

    if expected_frequencies is None:
        # Assume equal expected frequencies
        total_sum = np.sum(observed_frequencies)
        expected_frequencies = np.full_like(observed_frequencies, total_sum / len(observed_frequencies), dtype=float)
    else:
        expected_frequencies = np.asarray(expected_frequencies)
        if len(observed_frequencies) != len(expected_frequencies):
            raise ValueError("Observed and expected frequencies must have the same length.")
        if np.sum(observed_frequencies) != np.sum(expected_frequencies):
             print("Warning: Sum of observed frequencies does not equal sum of expected frequencies. Ensure consistency.")

    # Perform the Chi-square goodness-of-fit test
    chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). The observed distribution significantly differs from the expected distribution."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). The observed distribution does not significantly differ from the expected distribution."

    return chi2_statistic, p_value, interpretation

# Example Usage for Q14:
# Scenario: A die is rolled 100 times. Are the observed frequencies significantly different from a fair die?
observed_rolls_q14 = np.array([15, 18, 12, 16, 20, 19]) # Observed counts for faces 1-6
# Expected frequencies for a fair die (100 rolls / 6 faces)
expected_rolls_q14 = np.array([100/6, 100/6, 100/6, 100/6, 100/6, 100/6])

chi2_stat_q14, p_value_q14, interpretation_q14 = \
    chi_square_goodness_of_fit(observed_rolls_q14, expected_rolls_q14)

print(f"Q14 Chi-square Goodness-of-Fit Test:")
print(f"Chi-square Statistic: {chi2_stat_q14:.3f}")
print(f"P-Value: {p_value_q14:.3f}")
print(f"Interpretation: {interpretation_q14}\n")
```

```python
import numpy as np
import scipy.stats as stats

# Question 16: Implement an F-test using Python to compare the variances of two random samples.
def f_test_compare_variances(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two independent random samples.
    Args:
        sample1 (array-like): Data for the first sample.
        sample2 (array-like): Data for the second sample.
        alpha (float): Significance level.
    Returns:
        tuple: (f_statistic, p_value, interpretation_message)
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Ensure the larger variance is in the numerator to get F >= 1
    if var1 >= var2:
        f_statistic = var1 / var2
        df1 = len(sample1) - 1
        df2 = len(sample2) - 1
    else:
        f_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1

    # Calculate p-value (upper tail since F >= 1)
    p_value = stats.f.sf(f_statistic, df1, df2) * 2 # Multiply by 2 for a two-tailed test

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the variances of the two samples."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the variances of the two samples."

    return f_statistic, p_value, interpretation

# Example Usage for Q16:
np.random.seed(51)
sample1_q16 = np.random.normal(loc=10, scale=3, size=25)
sample2_q16 = np.random.normal(loc=12, scale=5, size=30) # Sample 2 has higher variance

f_stat_q16, p_value_q16, interpretation_q16 = f_test_compare_variances(sample1_q16, sample2_q16)
print(f"\nQ16 F-Test to Compare Variances:")
print(f"Variance of Sample 1: {np.var(sample1_q16, ddof=1):.3f}")
print(f"Variance of Sample 2: {np.var(sample2_q16, ddof=1):.3f}")
print(f"F-Statistic: {f_stat_q16:.3f}")
print(f"P-Value: {p_value_q16:.3f}")
print(f"Interpretation: {interpretation_q16}\n")
```

```python
import pandas as pd
import scipy.stats as stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Question 17: Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.
def perform_one_way_anova(data_df, value_col, group_col, alpha=0.05):
    """
    Performs a one-way ANOVA test and interprets the results.
    Args:
        data_df (pd.DataFrame): DataFrame containing the data.
        value_col (str): Name of the column with the dependent variable (numeric).
        group_col (str): Name of the column with the grouping variable (categorical).
        alpha (float): Significance level.
    Returns:
        tuple: (f_statistic, p_value, interpretation_message)
    """
    # Using scipy.stats.f_oneway for simplicity with separate arrays
    # A more robust approach uses statsmodels for full ANOVA table

    # Separate data into groups
    groups = [data_df[value_col][data_df[group_col] == g].values for g in data_df[group_col].unique()]

    f_statistic, p_value = stats.f_oneway(*groups)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the means of at least two of the groups."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the means of the groups."

    return f_statistic, p_value, interpretation

# Example Usage for Q17:
np.random.seed(52)
# Simulate data for three different teaching methods
data_q17 = pd.DataFrame({
    'Score': np.concatenate([
        np.random.normal(loc=75, scale=10, size=30), # Method A
        np.random.normal(loc=80, scale=10, size=30), # Method B
        np.random.normal(loc=70, scale=10, size=30)  # Method C
    ]),
    'Method': ['A']*30 + ['B']*30 + ['C']*30
})

f_stat_q17, p_value_q17, interpretation_q17 = \
    perform_one_way_anova(data_q17, 'Score', 'Method')

print(f"Q17 One-Way ANOVA Test:")
print(f"F-Statistic: {f_stat_q17:.3f}")
print(f"P-Value: {p_value_q17:.3f}")
print(f"Interpretation: {interpretation_q17}\n")
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Question 18: Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
def one_way_anova_plot_results(data_df, value_col, group_col, alpha=0.05):
    """
    Performs a one-way ANOVA test, interprets, and plots the results.
    Args:
        data_df (pd.DataFrame): DataFrame containing the data.
        value_col (str): Name of the column with the dependent variable (numeric).
        group_col (str): Name of the column with the grouping variable (categorical).
        alpha (float): Significance level.
    Returns:
        tuple: (f_statistic, p_value, interpretation_message)
    """
    # Perform ANOVA
    groups = [data_df[value_col][data_df[group_col] == g].values for g in data_df[group_col].unique()]
    f_statistic, p_value = stats.f_oneway(*groups)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the means of at least two of the groups."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the means of the groups."

    # Plotting results with boxplots
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_col, y=value_col, data=data_df, palette='viridis')
    plt.title(f'Comparison of {value_col} across {group_col} Groups')
    plt.xlabel(group_col)
    plt.ylabel(value_col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return f_statistic, p_value, interpretation

# Example Usage for Q18:
np.random.seed(53)
data_q18 = pd.DataFrame({
    'Yield': np.concatenate([
        np.random.normal(loc=50, scale=5, size=20), # Fertilizer A
        np.random.normal(loc=55, scale=5, size=20), # Fertilizer B
        np.random.normal(loc=48, scale=5, size=20)  # Fertilizer C
    ]),
    'Fertilizer': ['A']*20 + ['B']*20 + ['C']*20
})

f_stat_q18, p_value_q18, interpretation_q18 = \
    one_way_anova_plot_results(data_q18, 'Yield', 'Fertilizer')

print(f"Q18 One-Way ANOVA Test and Plot:")
print(f"F-Statistic: {f_stat_q18:.3f}")
print(f"P-Value: {p_value_q18:.3f}")
print(f"Interpretation: {interpretation_q18}\n")
```

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.gofplots as sm_gof

# Question 19: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
def check_anova_assumptions(data_df, value_col, group_col, alpha=0.05):
    """
    Checks the assumptions for ANOVA: Normality, Independence, and Homogeneity of Variances.
    Args:
        data_df (pd.DataFrame): DataFrame containing the data.
        value_col (str): Name of the column with the dependent variable (numeric).
        group_col (str): Name of the column with the grouping variable (categorical).
        alpha (float): Significance level for tests.
    """
    groups_data = [data_df[value_col][data_df[group_col] == g] for g in data_df[group_col].unique()]
    group_names = data_df[group_col].unique()

    print(f"Q19 Checking ANOVA Assumptions for '{value_col}' by '{group_col}':")

    # 1. Independence of Observations:
    print("\n1. Independence of Observations:")
    print("   - This assumption is primarily based on the experimental design.")
    print("   - Data points within and between groups should not influence each other.")
    print("   - Ensure random sampling and no systematic bias in data collection.")

    # 2. Normality of Residuals (or within each group):
    print("\n2. Normality Assumption (using Shapiro-Wilk Test and Q-Q plots for each group):")
    for i, group_data in enumerate(groups_data):
        shapiro_stat, shapiro_p = stats.shapiro(group_data)
        print(f"   - Group '{group_names[i]}': Shapiro-Wilk p-value = {shapiro_p:.3f}")
        if shapiro_p < alpha:
            print(f"     -> Suggests non-normality for group '{group_names[i]}' (p < {alpha}).")
        else:
            print(f"     -> Suggests normality for group '{group_names[i]}' (p >= {alpha}).")

        # Q-Q plot for visual inspection
        fig = sm_gof.qqplot(group_data, line='s', fit=True)
        plt.title(f'Q-Q Plot for Normality: Group "{group_names[i]}"')
        plt.show()

    # 3. Homogeneity of Variances (Homoscedasticity):
    print("\n3. Homogeneity of Variances (using Levene's Test):")
    # Levene's test is preferred over Bartlett's as it's less sensitive to non-normality
    levene_stat, levene_p = stats.levene(*groups_data)
    print(f"   - Levene's Test p-value = {levene_p:.3f}")
    if levene_p < alpha:
        print(f"     -> Reject H0: Variances are not equal (p < {alpha}). Consider Welch's ANOVA or transformations.")
    else:
        print(f"     -> Fail to reject H0: Variances are equal (p >= {alpha}). Assumption met.")

# Example Usage for Q19:
np.random.seed(54)
data_q19 = pd.DataFrame({
    'Measurement': np.concatenate([
        np.random.normal(loc=10, scale=2, size=30), # Group 1 (normal, low var)
        np.random.normal(loc=12, scale=3, size=30), # Group 2 (normal, medium var)
        np.random.normal(loc=15, scale=5, size=30)  # Group 3 (normal, high var - to potentially fail homoscedasticity)
    ]),
    'Category': ['A']*30 + ['B']*30 + ['C']*30
})

check_anova_assumptions(data_q19, 'Measurement', 'Category')

# Example with non-normal data (e.g., exponential) for demonstration of Shapiro-Wilk failure
print("\n--- Checking assumptions with intentionally non-normal data ---")
data_q19_non_normal = pd.DataFrame({
    'Measurement': np.concatenate([
        np.random.exponential(scale=1, size=30),
        np.random.exponential(scale=1.5, size=30)
    ]),
    'Category': ['X']*30 + ['Y']*30
})
check_anova_assumptions(data_q19_non_normal, 'Measurement', 'Category')
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Question 20: Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
def two_way_anova_visual(data_df, dependent_var, factor1, factor2, alpha=0.05):
    """
    Performs a two-way ANOVA test, interprets, and visualizes main and interaction effects.
    Args:
        data_df (pd.DataFrame): DataFrame containing the data.
        dependent_var (str): Name of the column with the dependent variable (numeric).
        factor1 (str): Name of the first categorical independent variable.
        factor2 (str): Name of the second categorical independent variable.
        alpha (float): Significance level.
    """
    # Create the OLS model formula
    formula = f'{dependent_var} ~ C({factor1}) + C({factor2}) + C({factor1}):C({factor2})'
    model = ols(formula, data=data_df).fit()

    # Perform ANOVA
    anova_table = sm.stats.anova_lm(model, typ=2) # Type 2 sums of squares recommended for unbalanced designs

    print(f"\nQ20 Two-Way ANOVA Results for {dependent_var} by {factor1} and {factor2}:")
    print(anova_table)

    # Interpretation
    print("\nInterpretation:")

    # Interaction Effect
    p_interaction = anova_table.loc[f'C({factor1}):C({factor2})', 'PR(>F)']
    if p_interaction < alpha:
        print(f" - Significant interaction effect (p={p_interaction:.3f} < alpha={alpha}).")
        print(f"   The effect of {factor1} on {dependent_var} depends on the level of {factor2}, and vice-versa.")
    else:
        print(f" - No significant interaction effect (p={p_interaction:.3f} >= alpha={alpha}).")
        print(f"   The effects of {factor1} and {factor2} on {dependent_var} are independent.")

    # Main Effects (interpret if interaction is NOT significant, or if significant, cautiously)
    p_factor1 = anova_table.loc[f'C({factor1})', 'PR(>F)']
    p_factor2 = anova_table.loc[f'C({factor2})', 'PR(>F)']

    if p_factor1 < alpha:
        print(f" - Significant main effect of {factor1} (p={p_factor1:.3f} < alpha={alpha}).")
    else:
        print(f" - No significant main effect of {factor1} (p={p_factor1:.3f} >= alpha={alpha}).")

    if p_factor2 < alpha:
        print(f" - Significant main effect of {factor2} (p={p_factor2:.3f} < alpha={alpha}).")
    else:
        print(f" - No significant main effect of {factor2} (p={p_factor2:.3f} >= alpha={alpha}).")

    # Visualization of Interaction Effect (Interaction Plot)
    plt.figure(figsize=(10, 7))
    sns.pointplot(x=factor1, y=dependent_var, hue=factor2, color='black', data=data_df, capsize=.1, errorbar='se', palette='viridis')
    plt.title(f'Interaction Plot: {dependent_var} by {factor1} and {factor2}')
    plt.xlabel(factor1)
    plt.ylabel(dependent_var)
    plt.grid(True)
    plt.show()

# Example Usage for Q20:
np.random.seed(55)
# Simulate data for drug efficacy based on Drug Type and Patient Gender
data_q20 = {
    'Drug_Effectiveness': np.random.normal(loc=50, scale=10, size=100),
    'Drug_Type': np.random.choice(['Drug A', 'Drug B'], size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100)
}
data_q20_df = pd.DataFrame(data_q20)

# Introduce a subtle interaction: Drug B works better for Females
data_q20_df.loc[(data_q20_df['Drug_Type'] == 'Drug B') & (data_q20_df['Gender'] == 'Female'), 'Drug_Effectiveness'] += 10
data_q20_df.loc[(data_q20_df['Drug_Type'] == 'Drug A') & (data_q20_df['Gender'] == 'Male'), 'Drug_Effectiveness'] -= 5


two_way_anova_visual(data_q20_df, 'Drug_Effectiveness', 'Drug_Type', 'Gender')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Question 21: Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.
def visualize_f_distribution(df_pairs=[(1, 1), (5, 2), (10, 10), (100, 100)]):
    """
    Visualizes the F-distribution for different pairs of degrees of freedom.
    Args:
        df_pairs (list of tuples): List of (df1, df2) pairs for numerator and denominator degrees of freedom.
    """
    plt.figure(figsize=(10, 6))
    x = np.linspace(0, 5, 500) # F-distribution often has its main mass near 1

    print("Q21 F-Distribution Characteristics and Use in Hypothesis Testing:")
    for df1, df2 in df_pairs:
        pdf = stats.f.pdf(x, df1, df2)
        plt.plot(x, pdf, label=f'df1={df1}, df2={df2}')
        print(f"- For df1={df1}, df2={df2}:")
        print(f"  - Characteristics: Non-negative, skewed to the right. The shape depends on both numerator (df1) and denominator (df2) degrees of freedom.")
        if df1 == 1 and df2 > 2:
             print(f"    - Note: F-distribution with df1=1 is related to the t-distribution (F = t^2).")

    print("\nUse in Hypothesis Testing:")
    print("- **ANOVA (Analysis of Variance):** The primary use is in ANOVA tests, where the F-statistic is the ratio of variance between groups to variance within groups. A large F-statistic suggests that group means are significantly different.")
    print("- **Comparing Two Variances:** The F-test can directly compare the variances of two populations (as shown in Q16), where the F-statistic is the ratio of their sample variances.")
    print("- **Regression Analysis:** Used to test the overall significance of a regression model (whether the model explains a significant portion of variance in the dependent variable) or to compare nested models.")

    plt.title('F-Distribution for Different Degrees of Freedom')
    plt.xlabel('F-Value')
    plt.ylabel('Probability Density Function')
    plt.legend()
    plt.grid(True)
    plt.ylim(bottom=0)
    plt.show()

# Example Usage for Q21:
visualize_f_distribution()
```

```python
import numpy as np
import scipy.stats as stats

# Question 23: Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
def simulate_and_test_means(num_groups=2, sample_size_per_group=30, true_means=None, std_dev=10, hypothesized_mean_diff=0, alpha=0.05):
    """
    Simulates data from normal distributions and performs hypothesis testing on means.
    Args:
        num_groups (int): Number of groups to simulate.
        sample_size_per_group (int): Number of data points per group.
        true_means (list, optional): List of true population means for each group.
                                     If None, uses 0, 0 for 2 groups; 0,1,2 for 3+ groups.
        std_dev (float): Standard deviation for data simulation.
        hypothesized_mean_diff (float): The hypothesized difference in means (for two-sample tests).
        alpha (float): Significance level.
    """
    if num_groups < 2:
        raise ValueError("At least two groups are needed for mean comparison.")

    if true_means is None:
        if num_groups == 2:
            true_means = [0, 0 + hypothesized_mean_diff] # Simulate under null by default unless diff specified
        else:
            true_means = list(range(num_groups)) # e.g., [0, 1, 2] for 3 groups

    simulated_groups_data = [np.random.normal(loc=mean, scale=std_dev, size=sample_size_per_group) for mean in true_means]

    print(f"\nQ23 Simulate Data and Evaluate Means (Number of Groups: {num_groups}, Sample Size: {sample_size_per_group})")
    for i, data in enumerate(simulated_groups_data):
        print(f"  Group {i+1} (True Mean={true_means[i]}): Sample Mean={np.mean(data):.3f}")

    if num_groups == 2:
        # Perform Independent Samples T-test
        t_statistic, p_value = stats.ttest_ind(simulated_groups_data[0], simulated_groups_data[1], equal_var=True)
        print(f"\nPerforming Independent Samples T-test:")
        print(f"  T-Statistic: {t_statistic:.3f}")
        print(f"  P-Value: {p_value:.3f}")
        if p_value < alpha:
            print(f"  Result: Reject H0 (p < {alpha}). Significant difference between group means.")
        else:
            print(f"  Result: Fail to reject H0 (p >= {alpha}). No significant difference between group means.")
    else:
        # Perform One-Way ANOVA
        f_statistic, p_value = stats.f_oneway(*simulated_groups_data)
        print(f"\nPerforming One-Way ANOVA Test:")
        print(f"  F-Statistic: {f_statistic:.3f}")
        print(f"  P-Value: {p_value:.3f}")
        if p_value < alpha:
            print(f"  Result: Reject H0 (p < {alpha}). At least one group mean is significantly different.")
        else:
            print(f"  Result: Fail to reject H0 (p >= {alpha}). No significant difference between group means.")

# Example Usage for Q23:
np.random.seed(57)
# Two groups, with a small true difference
simulate_and_test_means(num_groups=2, sample_size_per_group=40, true_means=[100, 105], std_dev=10)
# Three groups, with some true differences
simulate_and_test_means(num_groups=3, sample_size_per_group=30, true_means=[50, 52, 60], std_dev=8)
```

```python
import numpy as np
import scipy.stats as stats

# Question 24: Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
def chi_square_test_population_variance(sample_data, hypothesized_variance, alpha=0.05, tail='two-tailed'):
    """
    Performs a Chi-square test for a single population variance.
    Args:
        sample_data (array-like): The sample observations.
        hypothesized_variance (float): The hypothesized population variance (sigma_0^2).
        alpha (float): Significance level.
        tail (str): 'two-tailed', 'right-tailed', or 'left-tailed'.
    Returns:
        tuple: (chi2_statistic, p_value, interpretation_message)
    """
    n = len(sample_data)
    if n <= 1:
        raise ValueError("Sample size must be greater than 1 to calculate variance.")

    sample_variance = np.var(sample_data, ddof=1) # Use ddof=1 for sample variance

    # Calculate Chi-square test statistic
    chi2_statistic = (n - 1) * sample_variance / hypothesized_variance

    df = n - 1

    p_value = None
    interpretation = ""

    if tail == 'right-tailed':
        p_value = stats.chi2.sf(chi2_statistic, df)
        if p_value < alpha:
            interpretation = f"Reject H0 (p={p_value:.3f} < alpha={alpha}). Sample variance ({sample_variance:.3f}) is significantly GREATER than hypothesized variance ({hypothesized_variance:.3f})."
        else:
            interpretation = f"Fail to reject H0 (p={p_value:.3f} >= alpha={alpha}). Sample variance is not significantly greater."
    elif tail == 'left-tailed':
        p_value = stats.chi2.cdf(chi2_statistic, df)
        if p_value < alpha:
            interpretation = f"Reject H0 (p={p_value:.3f} < alpha={alpha}). Sample variance ({sample_variance:.3f}) is significantly LESS than hypothesized variance ({hypothesized_variance:.3f})."
        else:
            interpretation = f"Fail to reject H0 (p={p_value:.3f} >= alpha={alpha}). Sample variance is not significantly less."
    elif tail == 'two-tailed':
        # For two-tailed, take the min of the two tails and multiply by 2
        p_value_lower = stats.chi2.cdf(chi2_statistic, df)
        p_value_upper = stats.chi2.sf(chi2_statistic, df)
        p_value = 2 * min(p_value_lower, p_value_upper) # Account for both tails

        if p_value < alpha:
            interpretation = f"Reject H0 (p={p_value:.3f} < alpha={alpha}). Sample variance ({sample_variance:.3f}) is significantly DIFFERENT from hypothesized variance ({hypothesized_variance:.3f})."
        else:
            interpretation = f"Fail to reject H0 (p={p_value:.3f} >= alpha={alpha}). No significant difference in variance."
    else:
        raise ValueError("Tail must be 'two-tailed', 'right-tailed', or 'left-tailed'.")

    return chi2_statistic, p_value, interpretation

# Example Usage for Q24:
np.random.seed(58)
sample_data_q24 = np.random.normal(loc=0, scale=1.2, size=30) # True variance 1.44
hypothesized_variance_q24 = 1.0 # Test if variance is 1.0

chi2_stat_q24, p_value_q24, interpretation_q24 = \
    chi_square_test_population_variance(sample_data_q24, hypothesized_variance_q24, tail='two-tailed')

print(f"\nQ24 Chi-square Test for Population Variance:")
print(f"Sample Variance: {np.var(sample_data_q24, ddof=1):.3f}")
print(f"Hypothesized Variance: {hypothesized_variance_q24:.3f}")
print(f"Chi-square Statistic: {chi2_stat_q24:.3f}")
print(f"P-Value: {p_value_q24:.3f}")
print(f"Interpretation: {interpretation_q24}\n")

# Example for right-tailed test
np.random.seed(59)
sample_data_q24_right = np.random.normal(loc=0, scale=1.5, size=20) # True variance 2.25
hypothesized_variance_q24_right = 1.0
chi2_stat_q24_right, p_value_q24_right, interpretation_q24_right = \
    chi_square_test_population_variance(sample_data_q24_right, hypothesized_variance_q24_right, tail='right-tailed')
print(f"Q24 Chi-square Test (Right-Tailed) for Population Variance:")
print(f"Sample Variance: {np.var(sample_data_q24_right, ddof=1):.3f}")
print(f"Hypothesized Variance: {hypothesized_variance_q24_right:.3f}")
print(f"Chi-square Statistic: {chi2_stat_q24_right:.3f}")
print(f"P-Value: {p_value_q24_right:.3f}")
print(f"Interpretation: {interpretation_q24_right}\n")
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest

# Question 25: Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
def z_test_compare_proportions(count1, nobs1, count2, nobs2, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test for comparing two population proportions.
    Args:
        count1 (int): Number of successes in sample 1.
        nobs1 (int): Number of observations (total trials) in sample 1.
        count2 (int): Number of successes in sample 2.
        nobs2 (int): Number of observations (total trials) in sample 2.
        alpha (float): Significance level.
        alternative (str): 'two-sided', 'smaller' (for P1 < P2), 'larger' (for P1 > P2).
    Returns:
        tuple: (z_statistic, p_value, interpretation_message)
    """
    # Use statsmodels' built-in function
    z_statistic, p_value = proportions_ztest(
        count=[count1, count2],
        nobs=[nobs1, nobs2],
        alternative=alternative
    )

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the two proportions."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the two proportions."

    return z_statistic, p_value, interpretation

# Example Usage for Q25:
# Scenario: Compare the success rate of a new marketing campaign vs. an old one.
# Campaign A: 60 successes out of 100 trials
# Campaign B: 75 successes out of 120 trials

count1_q25 = 60
nobs1_q25 = 100
count2_q25 = 75
nobs2_q25 = 120

z_stat_q25, p_value_q25, interpretation_q25 = \
    z_test_compare_proportions(count1_q25, nobs1_q25, count2_q25, nobs2_q25, alternative='two-sided')

print(f"Q25 Z-Test for Comparing Proportions:")
print(f"Proportion 1: {count1_q25/nobs1_q25:.3f}")
print(f"Proportion 2: {count2_q25/nobs2_q25:.3f}")
print(f"Z-Statistic: {z_stat_q25:.3f}")
print(f"P-Value: {p_value_q25:.3f}")
print(f"Interpretation: {interpretation_q25}\n")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Question 26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
# This is a duplicate of Q16, providing the same function with added visualization.
def f_test_compare_variances_visual(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two independent random samples, interprets, and visualizes.
    Args:
        sample1 (array-like): Data for the first sample.
        sample2 (array-like): Data for the second sample.
        alpha (float): Significance level.
    Returns:
        tuple: (f_statistic, p_value, interpretation_message)
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Ensure the larger variance is in the numerator to get F >= 1
    if var1 >= var2:
        f_statistic = var1 / var2
        df1 = len(sample1) - 1
        df2 = len(sample2) - 1
        label1 = "Sample 1"
        label2 = "Sample 2"
    else:
        f_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1
        label1 = "Sample 2"
        label2 = "Sample 1" # Labels adjusted for plotting consistently

    p_value = stats.f.sf(f_statistic, df1, df2) * 2 # Multiply by 2 for a two-tailed test

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). There is a statistically significant difference between the variances of the two samples."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). There is no statistically significant difference between the variances of the two samples."

    # Visualization
    plt.figure(figsize=(10, 6))

    # Plot histograms to show distributions and spreads
    sns.histplot(sample1, color='blue', label=f'{label1} (Var={var1:.2f})', kde=True, stat='density', alpha=0.6)
    sns.histplot(sample2, color='orange', label=f'{label2} (Var={var2:.2f})', kde=True, stat='density', alpha=0.6)

    # Plot F-distribution and critical region (conceptual)
    x_f_dist = np.linspace(0, f_statistic * 2 if f_statistic > 1 else 2, 500) # Extend x-axis based on F-stat
    f_pdf = stats.f.pdf(x_f_dist, df1, df2)

    # Calculate critical F-value for visualization (two-tailed)
    critical_f_lower = stats.f.ppf(alpha / 2, df1, df2)
    critical_f_upper = stats.f.ppf(1 - alpha / 2, df1, df2)

    plt.figure(figsize=(10, 6))
    plt.plot(x_f_dist, f_pdf, color='green', label=f'F-distribution (df1={df1}, df2={df2})')
    plt.axvline(f_statistic, color='red', linestyle='--', label=f'Calculated F-stat: {f_statistic:.2f}')
    plt.axvline(critical_f_lower, color='purple', linestyle=':', label=f'Critical F: {critical_f_lower:.2f}, {critical_f_upper:.2f}')
    plt.axvline(critical_f_upper, color='purple', linestyle=':')

    # Shade rejection regions (conceptual, can be complex for both tails)
    x_reject_lower = x_f_dist[x_f_dist < critical_f_lower]
    plt.fill_between(x_reject_lower, 0, stats.f.pdf(x_reject_lower, df1, df2), color='red', alpha=0.3)
    x_reject_upper = x_f_dist[x_f_dist > critical_f_upper]
    plt.fill_between(x_reject_upper, 0, stats.f.pdf(x_reject_upper, df1, df2), color='red', alpha=0.3, label='Rejection Region')

    plt.title(f'F-Test for Comparing Variances (P-value: {p_value:.3f})')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()


    return f_statistic, p_value, interpretation

# Example Usage for Q26:
np.random.seed(60)
sample1_q26 = np.random.normal(loc=10, scale=4, size=35) # Variance ~16
sample2_q26 = np.random.normal(loc=12, scale=2, size=40) # Variance ~4

f_stat_q26, p_value_q26, interpretation_q26 = f_test_compare_variances_visual(sample1_q26, sample2_q26)
print(f"\nQ26 F-Test to Compare Variances with Visualization:")
print(f"Variance of Sample 1: {np.var(sample1_q26, ddof=1):.3f}")
print(f"Variance of Sample 2: {np.var(sample2_q26, ddof=1):.3f}")
print(f"F-Statistic: {f_stat_q26:.3f}")
print(f"P-Value: {p_value_q26:.3f}")
print(f"Interpretation: {interpretation_q26}\n")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Question 27: Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
# This is a duplicate of Q14, providing the same function with added data simulation for completeness.
def chi_square_goodness_of_fit_simulated(num_observations, expected_proportions, alpha=0.05, simulate_true_dist_diff=False):
    """
    Simulates data and performs a Chi-square goodness-of-fit test.
    Args:
        num_observations (int): Total number of observations to simulate.
        expected_proportions (array-like): The hypothesized proportions for each category. Must sum to 1.
        alpha (float): Significance level.
        simulate_true_dist_diff (bool): If True, slightly alters the true proportions for simulation
                                        to demonstrate a potential rejection.
    Returns:
        tuple: (chi2_statistic, p_value, interpretation_message, observed_frequencies, expected_frequencies)
    """
    num_categories = len(expected_proportions)

    # Simulate data based on a potentially different true distribution
    if simulate_true_dist_diff:
        # Create a slightly different 'true' distribution
        true_proportions = np.array(expected_proportions) * (1 + np.random.uniform(-0.1, 0.1, num_categories))
        true_proportions /= np.sum(true_proportions) # Normalize to 1
    else:
        true_proportions = np.array(expected_proportions)

    # Simulate observed frequencies
    observed_frequencies = np.random.multinomial(n=num_observations, pvals=true_proportions)

    # Calculate expected frequencies based on the *hypothesized* proportions
    expected_frequencies_calc = num_observations * np.array(expected_proportions)

    # Perform the Chi-square goodness-of-fit test
    chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies_calc)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p={p_value:.3f} < alpha={alpha}). The observed distribution significantly differs from the expected distribution."
    else:
        interpretation = f"Fail to reject the null hypothesis (p={p_value:.3f} >= alpha={alpha}). The observed distribution does not significantly differ from the expected distribution."

    print(f"\nQ27 Chi-square Goodness-of-Fit Test with Simulated Data:")
    print(f"Hypothesized Proportions: {np.round(expected_proportions, 3)}")
    print(f"Simulated True Proportions: {np.round(true_proportions, 3)}")
    print(f"Observed Frequencies: {observed_frequencies}")
    print(f"Expected Frequencies (calculated): {np.round(expected_frequencies_calc, 2)}")
    print(f"Chi-square Statistic: {chi2_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Interpretation: {interpretation}")

    # Optional: Visualization
    categories = [f'Cat {i+1}' for i in range(num_categories)]
    x_pos = np.arange(len(categories))

    width = 0.35
    fig, ax = plt.subplots(figsize=(10, 6))
    rects1 = ax.bar(x_pos - width/2, observed_frequencies, width, label='Observed', color='skyblue')
    rects2 = ax.bar(x_pos + width/2, expected_frequencies_calc, width, label='Expected', color='lightcoral')

    ax.set_ylabel('Frequency')
    ax.set_title('Observed vs. Expected Frequencies')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(categories)
    ax.legend()
    plt.show()

    return chi2_statistic, p_value, interpretation, observed_frequencies, expected_frequencies_calc

# Example Usage for Q27:
np.random.seed(61)
# Scenario 1: Simulate data where the null hypothesis (expected proportions) is likely true
print("--- Scenario 1: True distribution matches expected (expect to fail to reject H0) ---")
chi_square_goodness_of_fit_simulated(
    num_observations=200,
    expected_proportions=[0.25, 0.25, 0.25, 0.25],
    simulate_true_dist_diff=False # True distribution matches expected
)

np.random.seed(62)
# Scenario 2: Simulate data where the null hypothesis is likely false
print("\n--- Scenario 2: True distribution differs from expected (expect to reject H0) ---")
chi_square_goodness_of_fit_simulated(
    num_observations=200,
    expected_proportions=[0.20, 0.20, 0.30, 0.30],
    simulate_true_dist_diff=True # True distribution will be slightly different
)
```
